## 교재의 네 가지 행렬을 코드로 구현한다.

#### 1. 소행렬 (mainors mtarix)
i 번째 행, j 번째 열을 제외한 나머지의 행렬식을 원소로 한다.

```python
import numpy as np
import matplotlib.pyplot as plt

# Create 4x4 matrix
A = np.random.randn(4, 4)

# minors matrix
M = np.zeros((4, 4))
for i in range(len(A)):
  Ar = np.delete(A, i, axis = 0) # delete row
  for j in range(len(A)):
    Arr = np.delete(Ar, j, axis = 1) # delete col
    M[i, j] = np.linalg.det(Arr)

# plot
plt.imshow(M, cmap = 'gray')
```

#### 2. 격자행렬(grid matrix)
+1과 -1이 번갈아 들어가는 행렬





```python
# Grid matrix

G = np.zeros((4, 4))

for i in range(len(G)):
  for j in range(len(G)):
    G[i, j] = (-1)**(i+1+j+1)

plt.imshow(G, cmap = 'gray')
```

#### 3. 여인수행렬
소행렬과 격자행렬의 아다마르곱

```python
Cof = M*G

plt.imshow(Cof, cmap = 'gray')
```

#### 4. 수반행렬(adjugate matrix)


```python
ADJ = Cof.T

plt.imshow(ADJ, cmap = 'gray')
```

## 수반행렬과 np.linalg.inv()로 구한 역행렬 비교
역행렬은 수반행렬을 det(A)로 나누어 구한다.

```python
# Inverse matrix of A
Ai = np.linalg.inv(A)

# plot
fig, axs = plt.subplots(1, 2, figsize = (12, 6))
axs[0].imshow(Ai, cmap = 'gray')
axs[0].set_title('np.linalg.inv()')
axs[1].imshow(ADJ/np.linalg.det(A), cmap = 'gray')
axs[1].set_title('Computed Inverse matrix')
```
