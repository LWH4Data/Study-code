## 연습 문제 4 : 교재 풀이
1. 함수를 통해 소프트 코딩을 한 것이 인상적이다. 간단한 계산이 아니라 일반적으로 자주 사용하게 되는 것들은 함수로 만드는 것이 좋겠다.

```python
def addMatrices(A, B):

  # check that both matrics have same size
  if A.shape != B.shape:
    raise('Matrices must be the same size!')

  # initialize sum matrix
  C = np.zeros(A.shape)

  # sum!
  for i in range(A.shape[0]):
    for j in range(A.shape[1]):
      C[i, j] = A[i, j] + B[i, j]

  return C

# test the function
M1 = np.zeros((6, 4))
M2 = np.ones((6, 4))

addMatrices(M1, M2)
```

## 연습 문제 8 : 문제 풀이
1. 교재 풀이의 경우 오차 제곱합을 구현하고 해당 오차 제곱합이 일정 수치(10**-15)보다 작은가를 기준으로 boolean을 출력한다.
2. 나의 경우 np.isclose()를 활용하였다. 다만 이 경우 행렬의 형태로 출력된다.

```python
def isMatrixSymmetric(S):

  # difference between matrix and its transpose
  D = S-S.T

  # check whether sum of squared errors (SSE) is smaller than a threshold
  sse = np.sum(D**2)

  # output TRUE if sse is tiny; Flase means the matrix is a symmetric
  return sse < 10**-15

# note: There are many other ways you could solve this.
# If you to explore different methods, consider np.all() or np.isclose()
```

```python
# create symmetric and nonsymmetric matrices
A = np.random.randn(4, 4)
AtA = A.T@A

# Test!
print(isMatrixSymmetric(A))
print(isMatrixSymmetric(AtA))
```

## 연습 문제 9 : 직접 해보기
1. 비대칭 사각 행렬로부터 대칭 행렬을 만드는 뎃셈 방법.

```python
import numpy as np

# Make random matrix
A = np.random.randn(4, 4)

# 평균화
Ar = A.T + A

def isMatrixSymmetric(S):

  # difference between matrix and its transpose
  D = S-S.T

  # check whether sum of squared errors (SSE) is smaller than a threshold
  sse = np.sum(D**2)

  # output TRUE if sse is tiny; Flase means the matrix is a symmetric
  return sse < 10**-15

print(isMatrixSymmetric(Ar))
```

  - 두 번째 : 대각 원소가 1, 4 ,9 ,16인 대각 행렬
  - 세 번째 : 대각 원소가 두 번째 행렬의 제곱근인 대각 행렬

```python
import numpy as np

# Make matrix that the all elements are 1.
m1 = np.ones((4, 4))

# Second matrix
m2 = np.zeros((4, 4))
for i in range(0, 4):
  m2[i, i] = (i+1)**2

# Third matrix
m3 = np.zeros((4, 4))
for i in range(0, 4):
  m3[i, i] = i+1


# Multiply m1 * m2
print('m2 x m1 : \n', m2@m1, '\n')
print('m1 x m1 : \n', m1@m2, '\n')

# Multiply m1 & m3
print('m3 x m1 x m3 : \n', m3@m1@m3)
```

## 연습 문제 11 : 교재 풀이
테스트를 해보는 문제라 풀이 보다 아래 두 가지를 정리해야할 것 같다

> 1. 대각 행렬을 왼쪽에서 곱하면 오른쪽 행렬의 행이 해당 대각 원소 크기만큼 조겅된다.
2. 대각 행렬을 오른쪽에서 곱하면 왼쪽 행렬의 열이 해당 대각 원소 크기만큼 조정된다.

```python
n = 4

# create 'base' matrices
O = np.ones((n, n))
D = np.diag(np.arange(1, n+1)**2)
S = np.sqrt(D)

# pre- and post-multiply
pre = D@O
pst = O@D

# and both
both = S@O@S


# print outthe 'base matrices
print('Ones matrix:')
print(O), print(' ')

print('Diagonal matrix:')
print(D), print(' ')

print('Sqrt-diagonal matrix:')
print(S), print(' ')


print('Pre-multiply by diagonal:')
print(pre), print(' ')

print('Post-multiply by diagonal:')
print(pst), print(' ')

print('Pre- and post-multiply by sqrt-diagonal:')
print(both)
```
